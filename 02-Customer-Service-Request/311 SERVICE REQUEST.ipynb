```python
import pandas as pd
import matplotlib.pyplot as plt


df['Created Date'] = pd.to_datetime(df['Created Date'])
df['Closed Date'] = pd.to_datetime(df['Closed Date'])

df['Request_Closing_Time'] = df['Closed Date'] - df['Created Date']

percentile_value = 95  
closing_time_percentile = df['Request_Closing_Time'].quantile(percentile_value / 100)
average_closing_time = df[df['Request_Closing_Time'] <= closing_time_percentile]['Request_Closing_Time'].mean()


plt.figure(figsize=(8, 6))  
plt.bar('Overall', average_closing_time.total_seconds() / 3600, color='skyblue')
plt.title('Average Request Closing Time (Excluding Outliers)')
plt.xlabel('Category')
plt.ylabel('Time (hours)')
plt.tight_layout()

plt.show()
```
